```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mushrooms.csv")
df.head()
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
for col in df:
    print(df[col].value_counts(), '\n')
```

```python
df['class'] = df['class'].map({'e':'edible', 'p':'poisonous'})
df['odor'] = df['odor'].map({'a':'almond', 'l':'anise', 'c':'creosote', 'y':'fishy', 
                'm':'musty', 'n':'none', 'p':'pungent', 's':'spicy'})
```

```python
colours = ['#FFFF99','#ADFF2F','#00BFFF','#FA8072','#FFEBCD','#800000','#40E0D0','#808080','#2E8B57']

fig, ax = plt.subplots(figsize=(8, 5))
df.odor.value_counts().plot(kind = 'barh', 
                            color = colours);
ax.set_xlabel("Count");
ax.set_ylabel("Odor")
ax.set_title("Mushroom Odors");
```
